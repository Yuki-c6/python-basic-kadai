```python
"""
実践17-2:
pipでtqdmをインストールしてください。
for i in range(100000):を実行する際、tqdmを利用してください。内部処理では時間がかかる処理またはsleepを用いて毎回0.01sec程度停止するようにしてください。
"""
#tqdmモジュールをインポート
import tqdm

# numを初期化
num = 0

# numにループ回数を格納する
for i in tqdm.tqdm(range(100000)):
    num = i

print(num)

# numに0.01秒ごとに停止するループ回数を格納する
for i in tqdm.tqdm(range(1000)):
    time.sleep(0.01)
    num = i

print(num)
```

```python
"""
実践17-3:
標準ライブラリ time を用いて、時間がかかる処理を実装し、時間を計測してください。
"""
# timeモジュールをインポート
import time

# 開始時間を取得
start_time = time.time()

num = 0

# numに0.01秒ごとに停止するループ回数を格納する
for i in range(1000):
    time.sleep(0.01)
    num = i

# 処理終了時間を取得
end_time = time.time()

# 処理にかかる時間を計算し出力する
processing_time = end_time - start_time
print(f"処理時間{processing_time}秒")
print(num)
```

```python
"""
実践17-4:
メニュー番号と対応する料理名がdict形式で定義されているdish_infoがあります
リスト dish_orderにはメニュー番号が格納されています
この状況で、
(1) dish_orderからメニュー番号を取り出し
(2) dish_infoから対応する料理名を取得し
(3) 料理名を一つのリストに格納し、出力するコードを書いてください
"""
# 注文に対する料理名返すリストを定義
def get_order_names(dish_info: dict, dish_order: list):
    # 料理名を格納するリストを定義
    order_names: list[str] = []

    # 注文の数繰り返し処理を行う
    for order in dish_order:

        # dish_infoと合致するものの値をリストに追加していく
        for key, value in dish_info.items():
            if order == key:
                order_names.append(value)

    return order_names

dish_info = {1: "Curry", 2: "Pizza", 3: "Salad", 4: "Spaghetti"}
dish_order = [3, 1, 4, 1, 2]

# 関数を呼び出し出力する
print(get_order_names(dish_info, dish_order))
```

```python
"""
実践17-5: 従業員名簿の作成
会社の従業員名簿を作成するプログラムを作成してください。以下の仕様を満たすようにしてください。
(1) Employeeクラスを作成する
属性: name(名前)、age(年齢)、department(所属部署)、salary(給与)
__init__メソッド(コンストラクタ)で上記属性を初期化する
raise_salaryメソッドを作成し、self.salaryに5%アップした値を設定する
(2) 従業員のリストemployeesを作成する
(3) create_employee関数を作成する
引数としてname、age、department、salaryを受け取る
Employeeクラスのインスタンスを作成し、返す
(4) create_employee関数を使って、employeesリストに5人の従業員を追加する
適当な名前、年齢、部署、給与を設定する
(5) 全従業員の名簿を出力する
従業員一人につき、name、age、department、salaryを出力する
(6) 35歳以上の従業員に5%の給与アップを実施する
employeesリストから35歳以上の従業員を探し出す
その従業員インスタンスのraise_salaryメソッドを呼び出す
(7) 新しい従業員名簿を出力する(給与アップ後の名簿)
"""
# Employeeクラスを定義する
class Employee:
    # コンストラクタを定義する
    def __init__(self, name: str, age: int, department: str, salary: int):
        self.name = name
        self.age = age
        self.department = department
        self.salary = salary

    # 昇給メソッドを定義する
    def raise_salary(self, rate: int):
        self.salary = round(self.salary * (1 + rate / 100))

# 従業員リストを作成する
employees: list[str] = []

# create_employee関数を定義する
def create_employee(name: str, age: int, department: str, salary: int):
    return Employee(name, age, department, salary)

# ceate_employee関数でインスタンスを作成し、従業員リストに追加する
employees.append(create_employee("侍太郎", 26, "営業部",260000))
employees.append(create_employee("侍一郎", 28, "製造部",280000))
employees.append(create_employee("侍二郎", 30, "総務部",300000))
employees.append(create_employee("侍三郎", 32, "経理部",320000))
employees.append(create_employee("侍四郎", 35, "開発部",350000))
employees.append(create_employee("侍五郎", 37, "人事部",370000))
employees.append(create_employee("侍六郎", 42, "営業部",420000))

# 従業員名簿を出力する
for employee in employees:
    print(f"{employee.name}, {employee.age}, {employee.department}, {employee.salary}")

# 35歳以上にraise_salaryメソッドを昇給率5％で適用する
for employee in employees:
    if employee.age >= 35:
        employee.raise_salary(5)

# 昇給後の従業員名簿を出力する
print("以下改定従業員名簿")
for employee in employees:
    print(f"{employee.name}, {employee.age}, {employee.department}, {employee.salary}")
```

```python
"""
実践17-6
問題: ユーザプロフィールクラスの作成
この問題では、ユーザプロフィールを表すクラスを作成し、さらにそのクラスを加工するための関数を実装します。

UserProfileクラスを作成してください。このクラスには以下のプロパティを持たせます。
name (名前)
age (年齢)
email (メールアドレス)
bio (自己紹介文)

update_profile関数を作成してください。この関数はUserProfileインスタンスを引数として受け取り、以下の処理を行います。
nameプロパティを大文字に変換する
ageプロパティに5を加算する
emailプロパティに"@example.com"を追加する (既にドメインが含まれている場合は変更しない)
bioプロパティが空文字列の場合は"No bio provided."に設定する

update_profile関数から返されたUserProfileインスタンスを使って、以下のように出力してください。
名前: [名前]
年齢: [年齢]
メールアドレス: [メールアドレス]
自己紹介: [自己紹介文]

次のユーザプロフィールを使ってプログラムを実行し、結果を確認してください。
user1 = UserProfile("John Doe", 30, "john@example.com", "I am a software engineer.")
user2 = UserProfile("jane_smith", 25, "janesmith", "")
"""
# UserProfileクラスを定義
class UserProfile:
    # コンストラクタを定義
    def __init__(self, name: str, age: int, email: str, bio: str):
        self.name = name
        self.age = age
        self.email = email
        self.bio = bio

    # 出力用のメソッドを定義
    def display_info(self):
        print(f"名前:[{self.name}]")
        print(f"年齢:[{self.age}]")
        print(f"メールアドレス:[{self.email}]")
        print(f"自己紹介:[{self.bio}]")

# user_update関数を定義
def update_profile(user):
    # nameを大文字に変換する
    user.name = user.name.upper()

    # 年齢に5を加算する
    user.age = user.age + 5

    # メールアドレスにドメインが含まれない場合に追加する
    if  "@" not in user.email:
        user.email = user.email + "@example.com"

    # 自己紹介文が空白の場合、メッセージを返す
    if user.bio == "":
        user.bio = "No bio provided."


# インスタンス化を行う
user1 = UserProfile("John Doe", 30, "john@example.com", "I am a software engineer.")
user2 = UserProfile("jane_smith", 25, "janesmith","")

# 情報を更新したものを出力する
update_profile(user1)
user1.display_info()

update_profile(user2)
user2.display_info()
```

```python
"""
総合問題17-7: 日付計算
現在日から以下の条件を満たす日付を計算し、それぞれの日付を計算してください。

2週間後の日付
一カ月前の日付
365日後の日付
上記のいずれかを文字列として変数に格納してください。
"""
# datetimeモジュールをインポート
import datetime
from dateutil.relativedelta import relativedelta

# 現在時刻を取得する
now = datetime.datetime.now()

# 日数差をそれぞれ変数に格納する
td_2w = datetime.timedelta(weeks=2)
td_1m = relativedelta(months=1)
td_365d = datetime.timedelta(days=365)

# 現在との日数差を変数に格納する
after_2week = now + td_2w
before_1month = now - td_1m
after_365day = now + td_365d

print(after_2week.strftime("%Y/%m/%d"))
print(before_1month.strftime("%Y/%m/%d"))
print(after_365day.strftime("%Y/%m/%d"))

```

```python
"""
総合問題17-8(関数とクラスの応用): 関数渡し
関数の引数やクラスのメンバー(属性)に関数を渡すことができます。
クラスのインスタンス化時に関数を渡すことで、インスタンスごとに異なる動作を実現できます。
以下、問題です。

動物の鳴き声を出力するプログラムを作ってください。
しかし、動物の種類によって鳴き声は異なるので、動物ごとに鳴き声を出力する関数を用意するのは面倒です。そこで、動物クラスのインスタンスを生成する際に、鳴き声を出力する関数を渡せるようにしたいと思います。

Animalクラスを作成してください
初期化時に鳴き声関数を引数で受け取り属性として設定してください
make_soundメソッド内で鳴き声を鳴らしてください

"""
# Animal クラスを定義
class Animal:
    # コンストラクタを定義
    def __init__(self,sound):
        self.sound = sound

    # 鳴き声を出力するメソッドを定義
    def make_sound(self) -> str:
        print(self.sound)

# 鳴き声を返す関数を定義
def cry_sound(species: str) -> str:
    if species == "dog":
        return "bowwow"
    elif species == "cat":
        return "meow"
    else:
        return "・・・"

# インスタンスを定義
dog = Animal(cry_sound("dog"))
cat = Animal(cry_sound("cat"))

# 鳴き声を出力
dog.make_sound()
cat.make_sound()
```
