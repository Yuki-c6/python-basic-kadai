```python
# Humanクラスを定義
class Human:
    # 属性を定義
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    # 大人かどうか判別するメソッドを定義
    def checkadult(self):
        if self.age > 20:
            print("大人です")
        else:
            print("大人ではありません")

# インスタンスを格納するリストを定義
humans = []

# インスタンス化を行う
human_a = Human("侍太郎", 8)
humans.append(human_a)
human_b = Human("侍一郎", 18)
humans.append(human_b)
human_c = Human("侍二郎", 22)
humans.append(human_c)
human_d = Human("侍三郎", 35)
humans.append(human_d)

# 大人か判断するメソッドを実行
for human in humans:
    human.checkadult()
```
