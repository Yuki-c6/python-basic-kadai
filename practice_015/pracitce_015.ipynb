```python
"""
実践15-1:
Personクラスを定義し、以下の条件を満たすようにしてください。
コンストラクタでは、名前(name)と年齢(age)を受け取る
greet()メソッドを定義し、"Hello, my name is [名前]"と出力する
get_year_of_birth()メソッドを定義し、現在年(2024年)から年齢を引いた値を返す
"""
# datetimeモジュールをインポート
import datetime

# Personクラスを定義
class Person:
    # 属性を定義する
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    # 挨拶メソッドを定義する
    def greet(self):
        print(f"Hello, my name is {self.name}")

    # 誕生年を出力するメソッドを定義する
    def get_year_of_birth(self) -> int:
        date_time = datetime.datetime.now()
        year_of_birth: int = date_time.year - self.age
        return year_of_birth

# インスタンス化する
person = Person("侍太郎",29)

# メソッドを実行する
person.greet()
print(person.get_year_of_birth())

```

```python
"""
実践15-2:
BankAccountクラスを定義し、以下の条件を満たすようにしてください。
コンストラクタでは、口座番号(account_number)と残高(balance)を受け取る
deposit(amount)メソッドを定義し、指定された金額を残高に加える
withdraw(amount)メソッドを定義し、指定された金額を残高から引く(残高が不足している場合は"Insufficient balance"と出力する)
get_balance()メソッドを定義し、現在の残高を返す
"""
# BankAccountクラスを定義する
class BankAccount:
    # 属性を定義する
    def __init__(self, account_number: str, balance: int):
        self.account_number = account_number
        self.balance = balance

    # 入金メソッドを定義する
    def deposit(self, amount: int):
        self.balance = self.balance + amount

    # 出金メソッドを定義する
    def withdraw(self, amount):
        calculated_balance: int = self.balance - amount
        if calculated_balance < 0:
            print("Insufficient balance")
        else:
            self.balance = calculated_balance

    # 残高確認メソッドを定義する
    def get_balance(self) -> int:
        return self.balance

# インスタンス化する
bank_account = BankAccount("000-333-555-444", 10000)

# メソッドを実行する
bank_account.deposit(5000)
bank_account.withdraw(20000)
print(bank_account.get_balance())
```

```python
"""
実践15-3:
Rectangleという以下のクラスを定義し、メソッドを呼び出してください。
コンストラクタでは、幅(width)と高さ(height)を受け取る
calculate_area()メソッドを定義し、長方形の面積を計算して返す
calculate_perimeter()メソッドを定義し、長方形の周囲長を計算して返す
double_size()メソッドを定義し、幅と高さを2倍にする
"""
# Rectangleクラスを定義
class Rectangle:
    # 属性を定義する
    def __init__(self, width: int, height: int):
        self.width = width
        self.height = height

    # 面積を計算するメソッドを定義
    def calculate_area(self) -> int:
        area: int = self.width * self.height
        return area

    # 周囲長を計算するメソッドを定義
    def calculate_perimeter(self) -> int:
        perimeter: int = (self.width + self.height) * 2
        return perimeter

    # サイズを2倍にするメソッドを定義
    def double_size(self):
        self.width = self.width * 2
        self.height = self.height * 2

# インスタンス化する
rectangle = Rectangle(5, 10)

# メソッドを呼び出す
print(rectangle.calculate_area())
print(rectangle.calculate_perimeter())
rectangle.double_size()
print(rectangle.calculate_area())
print(rectangle.calculate_perimeter())
```

```python
"""
実践15-4:
Studentクラスを定義し、以下の条件を満たすようにしてください。
コンストラクタでは、名前(name)と年齢(age)を受け取る
add_grade(subject, grade)メソッドを定義し、受け取った科目(subject)と成績(grade)をdictionary形式で保存する
get_grades()メソッドを定義し、保存された全ての科目と成績をdictionaryとして返す
get_average_grade()メソッドを定義し、全ての科目の平均点を計算して返す
"""
# Studentクラスを定義する
class Student:
    # 属性を定義する
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age
        self.grade_data: dict = {}

    # 成績追加メソッドを定義する
    def add_grade(self, subject: str, grade: int):
        self.grade_data[subject] = grade

    # 保存されている成績を返すメソッドを定義する
    def get_grade(self) -> dict:
        return self.grade_data

    # 全ての科目の平均値を返すメソッドを定義する
    def get_average_grade(self) -> float:
        total_grade: int = sum(self.grade_data.values())
        grade_count: int = len(self.grade_data)
        grade_average: float = total_grade / grade_count
        return grade_average

# インスタンス化する
student = Student("侍太郎", 18)

# メソッドを呼び出す
student.add_grade("math", 90)
print(student.get_grade())
print(student.get_average_grade())

student.add_grade("english", 70)
student.add_grade("science", 85)
print(student.get_grade())
print(student.get_average_grade())
```
