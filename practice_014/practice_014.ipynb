```python
"""
実践1:次のコードでは、エラーが発生します。原因を説明し、修正した上で正しく動作するようにコードを書き換えてください。
原因:変数even_lstは関数my_func内で定義されたものであり、関数外ではエラーが発生する。
解決策: my_funcの返り値としてeven_lstが返されているため、my_funxの結果をresultに代入し出力する。
"""
def my_func(lst):
    even_lst = []
    for num in lst:
        if num % 2 == 0:
            even_lst.append(num)
    return even_lst

my_list = [1, 2, 3, 4, 5]
result = my_func(my_list)
print(f"Original list: {my_list}")
print(f"Even numbers: {result}")
```

```python
"""
実践2: 次のコードでは、グローバル変数とローカル変数の両方を使用しています。出力はどうなるでしょうか?結果を説明してください。また、コードはコピーせず手で打ち込んでください。
結果:グローバル変数で定義したxとローカル変数で定義したxは別物として扱われるため、my_func内ではローカル変数で定義したx=20が使用されるが
     関数の外では、最初に定義されたx=10が更新されずに使用される。
"""
x = 10

def my_func():
    x = 20
    print(f"Inside function: x = {x}")

my_func()
print(f"Outside function: x = {x}")
```
