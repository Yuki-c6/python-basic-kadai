```python
"""
実践1: 1から20までの数字を出力するプログラムを作成してください。ただし、5の倍数の場合は'Fizz'と出力してください。
"""
for num in range(1, 21):
    if num % 5 == 0:
        print("Fizz")
        continue

    print(num)
```

```python
"""
実践2: リストの要素を1つずつ表示するプログラムを作成してください。ただし、要素が'q'の場合は繰り返しを終了してください。
"""
unique: list[str] = ["u", "n", "i", "q", "u", "e"]

for value in unique:
    if value == "q":
        print("要素がqなので繰り返しを終了します")
        break

    print(value)

```

```python
"""
実践3: 1から100までの数字を出力するプログラムを作成してください。ただし、3の倍数の場合は'Fizz'、5の倍数の場合は'Buzz'、3と5の公倍数の場合は'FizzBuzz'と出力してください。
"""
for num in range(1,101):
    if num % 15 == 0:
        print("FizzBuzz")
        continue

    if num % 5 == 0:
        print("Buzz")
        continue

    if num % 3 == 0:
        print("Fizz")
        continue

    print(num)
```

```python
"""
実践4: 文字列(英語)を入力として受け取り、その文字列内の母音の数を数えるプログラムを作成してください。ただし、大文字と小文字は区別しないものとします。
"""
alphabets: str = input("英語で入力してください")
vowels: tuple[str]  = ("a", "e", "i", "o", "u", "A", "E", "I", "O", "U")
count: int  = 0

for alphabet in alphabets:
    if alphabet in vowels:
        count = count + 1

if count == 0:
    print("母音が含まれていません")
else:
    print(f"母音は{count}文字です")
```

```python
"""
実践5: 2つのリストを作成し、forを重ねて実行してください。その結果をprintしてください。
"""
columns: list[str] = ["A", "B", "C", "D", "E"]
rows: list[int] = [1, 2, 3, 4, 5]

for column in columns:
    for row in rows:
        print(f"{column}列{row}行目")
```

```python
"""
実践6: 2つのリストをforで回し、それらをtupleとして新しいリストへ格納するプログラムを作成してください。
list内が[(str, int),(str, int)]の場合どのような型ヒントを記載すればよいのか
"""

columns: list[str] = ["A", "B", "C", "D", "E"]
rows: list[int] = [1, 2, 3, 4, 5]
index: list = []


for column in columns:
    for row in rows:
        index.append((column, row))

print(index)
```

```python
"""
実践7: リストに数字を複数入力し、入力した数字の合計値が100を超えたら繰り返しを終了するプログラムを作成してください。inputとint変換を使うとわかりやすいです。
"""
input_nums: list[int] = []
sum = 0
input_nums.append(int(input("数値を入力してください")))

for num in input_nums:
    sum = sum + num

    if sum > 100:
        print(f"合計{sum}")
        print("合計値が100を超えたので繰り返しを終了します")
        break

    input_nums.append(int(input("数値を入力してください")))

print(input_nums)


```

```python
"""
実践8: 与えられたディクショナリーから、値が偶数のキーと値のペアのみを新しいディクショナリーとして出力するプログラムを作成してください。
"""
all_data: dict[int, str] = {1: "pen", 2: "desk", 3: "camera", 4: "phone", 5: "watch"}
select_data: dict[int, str] ={}

for key, value in all_data.items():
    if key % 2 == 0:
        select_data[key] = value

print(select_data)

```

```python
"""
実践9: 以下の応用コードを実行し動作を確認してください。
"""
user_names : list = ["侍太郎", "侍一郎", "侍二郎", "侍三郎", "侍四郎"]

# 配列 user_namesのインデックスと値を1つずつ順番に出力する
for index, value in enumerate(user_names):
    print(f"{index} : {value}")
```

```python
# そのほか：2重ループ
list_english = ["a", "b", "c"]
list_num = [1, 2, 3]

for l1 in list_english:
    for l2 in list_num:
        print(f"{l1}_{l2}")
```

```python
# そのほか：２つのリストの同時処理
list_english = [ "a", "b", "c"]
list_num = [1, 2, 3]

for l1, l2 in zip(list_english, list_num):
    print(f"{l1}_{l2}")
```
